**Problem set up**:

A lecture shall be finished in time interval $T$. The lecturer spread $k$ information in $n$ bits, where the redundancy $n-k$ is useful for missed times, charactered by noise level $\delta$. THe remained $T-n$ bits is for casual study or fun facts, which make the lecture more interesting, reduce noise, and improve decoding capability for students. From the student side, they need to decode out the $k$ logical information from $n$ noisy bits, while the $T-n$ bits are supposed to be easily distinguished.

general definition.
let $T=45$ since there are usually 45 mins in a lecture

```python
def p0(n,x):
    return (1-x)**n  # probability of no error
def p1(n,x):
    return n*x*(1-x)**(n-1)  # probability of one error
def p2(n,x):
    return n*(n-1)*(x**2)*(1-x)**(n-2)/2  # probability of two errors
def p3(n,x):
    return n*(n-1)*(n-2)*(x**3)*(1-x)**(n-3)/6  # probability of three errors
def p4(n,x):
    return n*(n-1)*(n-2)*(n-3)*(x**4)*(1-x)**(n-4)/24  # probability of four errors
ps=[p0,p1,p2,p3,p4]
def logical_error_rate(n,d,delta):
    assert d<9 # 'd must be less than 9'
    x=delta
    t=int(d/2)  # maximum number of errors to consider
    y=0
    for i in range(t):
        y += ps[i](n, x)  # accumulate probabilities for no error to t errors
    #y=(1-x)**n + n*x*(1-x)**(n-1) + n*(n-1)*(x**2)*(1-x)**(n-2)/2 #probability of no error or one error or two errors
    P=1-y # probability of logical error
    return P
P=logical_error_rate(n,d,delta)
print(f'Logical error rate for n={n}, d={d}, delta={delta} is {P}')
#print(f'Logical error rate for delta={delta} is {P}')
```
